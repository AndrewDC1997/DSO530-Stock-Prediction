```python
from os import listdir
from os.path import isfile, join

import os
import pandas as pd

from sklearn import preprocessing
import numpy as np

import keras
import tensorflow as tf
from keras.models import Model
from keras.layers import Dense, Dropout, LSTM, Input, Activation, concatenate
from keras import optimizers

import matplotlib.pyplot as plt

capi_list = []
fund_list = []
invest_num = []
# Set paths
p = os.path.abspath('..')
p1 = p + '\\raw data 17-19\\'
p2 = p + '\\engineered features 17-19\\'

# Get all file names.csv
onlyfiles = [f for f in listdir(p1) if isfile(join(p1, f))]

for f, stock_num in zip(onlyfiles[0:50], range(50)):
    
    # Read file
    raw_df = pd.read_csv(p1+f)
    tech_df = pd.read_csv(p2+f)
    raw_df = raw_df.drop('Time', axis=1)
    raw_df = raw_df.drop('Volume', axis=1)
    

#     tech_df = tech_df.drop('OBVlag', axis=1)
#     tech_df = tech_df.drop('ADlag', axis=1)
    tech_df = tech_df.drop('Unnamed: 0', axis=1)
    
    tech_df.index = tech_df.index+15
    tech_df = tech_df.reindex(np.arange(740+15))
    
    raw_df = pd.concat([raw_df, tech_df], axis = 1)
    raw_df = raw_df.drop('MA15lag', axis=1)
    raw_df = raw_df.drop('day15Returnlag', axis=1)
    

    # Parameters
    history_points = 15
    tech_elim_days = 16
    model_unseen_split_point = 504
    train_test_split_ratio = 0.8

    # Split data
    raw_df_model = raw_df[0:model_unseen_split_point]  # Size of 504
    raw_df_unseen = raw_df[model_unseen_split_point:756] # Size of 756-504

#     train_test_split_point = int(raw_df_model.shape[0] * train_test_split_ratio) # 403
#     raw_df_model_train = raw_df_model[:train_test_split_point] # Size of 504*0.8
#     raw_df_model_test = raw_df_model[train_test_split_point:] # Size of 504*0.2

    # Normalize
    data_normaliser = preprocessing.MinMaxScaler().fit(raw_df_model) # Use training data to fit normaliser

    data_normalised = data_normaliser.transform(raw_df)
    model_normalised = data_normaliser.transform(raw_df_model)  # Size = 504
    unseen_normalised = data_normaliser.transform(raw_df_unseen)  # Size = 756-504, unseen, only for final evaluation
    
    y_normaliser = preprocessing.MinMaxScaler()
    y_normaliser.fit(np.expand_dims(raw_df_model.loc[:, 'Close'].values, -1))


    # TOMORROW
    # X, Y for train and test (Model use)
    x_train_normalised = np.array([model_normalised[i  : i + history_points].copy() \
                                          for i in range(tech_elim_days, len(raw_df_model) - history_points)])
    y_train_normalised = np.array([model_normalised[:,3][i + history_points].copy() \
                                          for i in range(tech_elim_days, len(raw_df_model) - history_points)])
    y_train_normalised = np.expand_dims(y_train_normalised, -1)

    x_test_normalised = np.array([data_normalised[i  : i + history_points].copy() \
                        for i in range(tech_elim_days + x_train_normalised.shape[0], len(raw_df) - history_points)])
    y_test_normalised = np.array([data_normalised[:,3][i + history_points].copy() \
                        for i in range(tech_elim_days + x_train_normalised.shape[0], len(raw_df) - history_points)])
    y_test_normalised = np.expand_dims(y_test_normalised, -1)

#     unseen_x_normalised = np.array([data_normalised[i  : i + history_points].copy() \
#                           for i in range(model_unseen_split_point - history_points, len(data_normalised) - history_points)])
#     unseen_y_normalised = np.array([data_normalised[:,3][i + history_points].copy() \
#                           for i in range(model_unseen_split_point - history_points, len(data_normalised) - history_points)])
#     unseen_y_normalised = np.expand_dims(unseen_y_normalised, -1)

    # Unormalised data
    y_test_unscale = np.array([np.array(raw_df)[:,3][i + history_points].copy() \
                        for i in range(tech_elim_days + x_train_normalised.shape[0], len(raw_df) - history_points)])
    y_test_unscale = np.expand_dims(y_test_unscale, -1)
#     unseen_y_unscale = np.array([np.array(raw_df)[:,3][i + history_points].copy() \
#                           for i in range(model_unseen_split_point - history_points, len(data_normalised) - history_points)])
#     unseen_y_unscale = np.expand_dims(unseen_y_unscale, -1)


    # Setting Seeds
    np.random.seed(4)
    tf.random.set_seed(4)

    # Model Structures
    lstm_input = keras.layers.Input(shape=(history_points, len(raw_df.columns)), name='lstm_input')
    x = LSTM(50, name='lstm_0')(lstm_input)
    x = Dropout(0.2, name='lstm_dropout_0')(x)
    x = Dense(64, name='dense_0')(x)
    x = Activation('sigmoid', name='sigmoid_0')(x)
    x = Dense(1, name='dense_1')(x)
    output = Activation('linear', name='linear_output')(x)

    # for tomorrow
    model = Model(inputs=lstm_input, outputs=output)
    adam = optimizers.Adam(lr=0.0005)
    model.compile(optimizer=adam, loss='mae')



    # Train the model
    hist = model.fit(x=x_train_normalised, y=y_train_normalised, 
              batch_size=32, epochs=300, verbose=0, shuffle=True, validation_split=0.1)
#     evaluation = model.evaluate(x_test_normalised, y_test_normalised)
#     print(evaluation)



    # Predict on tomorrow test set
    y_test_predicted = model.predict(x_test_normalised)
    y_test_predicted = y_normaliser.inverse_transform(y_test_predicted)

#     y_test_predicted_temp = []
#     temp = []
#     for i in range(len(y_test_predicted)):
#         a = [0]*14
#         a = a.append(y_test_predicted[i][0])
#         y_test_predicted_temp.append(a)

#     y_test_predicted_temp = np.array(y_test_predicted_temp)
    
#     for i in range(len(y_test_predicted)):
#         y_test_predicted[i][0] = y_test_predicted_temp[i][-1]

    
    
    # Calculate the MAE
    real_mse = np.mean(abs(y_test_unscale - y_test_predicted))
    scaled_mse = real_mse / (np.max(y_test_unscale) - np.min(y_test_unscale)) * 100
    print(f'STOCK #{stock_num} Test mae: {round(real_mse, 4)} Percentage mae: {round(scaled_mse, 4)}')

    
    # Draw on test data
    plt.gcf().set_size_inches(10, 6, forward=True)
    
    start = 0
    end = -1

    real = plt.plot(np.arange(model_unseen_split_point, 756, 1), y_test_unscale, label='real')
    pred = plt.plot(np.arange(model_unseen_split_point, 756, 1), 
                    y_test_predicted, label='predicted tomo')

    plt.legend(['Real', 'Predicted'])
    plt.title(f'Test of stock#{stock_num}')
    plt.show()
    
    

#     if scaled_mse >=8:
#         continue
#     invest_num.append(stock_num)

    # Evaluation 1 earnings
    capi = 20000
    hold = 0
    buy_price = 0
    pnl = np.zeros(y_test_predicted.shape[0]-1)

    fund = []
    last_fund = 0
    for i in range(y_test_predicted.shape[0]):
        fund.append(capi + hold*y_test_unscale[i][0])

        # first day
        if i == 0:
            last_fund = capi + hold*y_test_unscale[i][0]
            continue

        # other day
        elif i != 251:
            pnl[i-1] = (capi + hold*y_test_unscale[i][0])/last_fund - 1
            
            # holding
            if hold:
                # take profit
                if ((y_test_predicted[i+1]-buy_price)/buy_price >= 0.30) | \
                   ((y_test_unscale[i]-buy_price)/buy_price >= 0.30):

                    capi += (hold*y_test_unscale[i][0])*(1-0.0065)
                    hold = 0

                # stop loss
                elif  ((y_test_predicted[i+1]-buy_price)/buy_price <= -0.03) | \
                      ((y_test_unscale[i]-buy_price)/buy_price <= -0.03):

                    capi += (hold*y_test_unscale[i][0])*(1-0.0065)
                    hold = 0

            # not holding
            else:
                # buy
                if ((y_test_predicted[i+1] > y_test_unscale[i]) & (y_test_predicted[i+1] > y_test_predicted[i])) \
                    & (tech_df['fastKlag'][model_unseen_split_point+i] > tech_df['fastDlag'][model_unseen_split_point+i]):
                    buy_price = y_test_unscale[i]
                    hold = capi/(y_test_unscale[i][0]*1.0065)
                    capi = 0


        # last day
        else:
            pnl[i-1] = (capi + hold*y_test_unscale[i])/last_fund - 1
            capi += hold*y_test_unscale[i][0]
            hold = 0
            
        last_fund = capi + hold*y_test_unscale[i]
                
    Sharp_ratio = np.sqrt(252) * np.mean(pnl) / np.std(pnl)            
    print('Earning: ', capi/20000)
    print('\n\n')
    #print('Sharp Ratio: ', Sharp_ratio)
    #print('Days making Profit: ', np.sum(pnl > 0))
    
    capi_list.append(capi)
    fund_list.append(fund)

fund_list = np.sum(fund_list, axis=0)
for i in range(len(fund_list[1:])):
    pnl[i] = fund_list[i+1]/fund_list[i] - 1
Sharp_ratio = np.sqrt(252) * np.mean(pnl) / np.std(pnl)
print('Earning: ', fund_list[-1] - 1000000)
print('Sharp Ratio: ', Sharp_ratio)
print('Days making Profit: ', np.sum(pnl > 0)) 

#capi_list                    
#print(capi_list)
#print(sum(capi_list))
#print(1000*len(invest_num))
```
